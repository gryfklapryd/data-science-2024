```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv"
df = pd.read_csv(url)

print("head: \n", df.head())
print("info: \n", df.info())

print("per kolom: \n", df.isnull().sum())

print("proporsi %: \n", df.isnull().mean())

print("selamat: \n", df[df['Survived']==1])
print("wanita: \n", df[df['Sex']=='female'])

print("kelas dengan tingkat keselamatan tertinggi : \n", df.groupby('Pclass')['Survived'].mean())
```
